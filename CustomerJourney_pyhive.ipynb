# 1. Does Gender have any impact on booking conversions?

```python


pd.read_sql('''select genderid,count(genderid) count_gender ,purchase_any  
               from travel_data join travel_data_demos on travel_data.userid = travel_data_demos.userid 
                where purchase_any = 1 and genderid in (1,2) group by genderid, purchase_any limit 10''',con=conn)
```

    We can see females have the highest number for booking conversions

# 2. Trend of booking conversions based on gross-income & life stages of consumers.

```python
pd.read_sql('''select purchase_any, BAS_bruto_jaarinkomen, SPSS_Lifestage, count(SPSS_Lifestage) as spss_count 
from travel_data join travel_data_demos on travel_data.userid = travel_data_demos.userid 
group by purchase_any, BAS_bruto_jaarinkomen, SPSS_Lifestage 
having purchase_any = 1 order by spss_count desc limit 10''', con=conn)
```

    For an income of:-  1: <€12.900 (minimum), Young Families are the highest, followed by Mature Singles

                    2: €12.900<=27.000 (below average), Empty nesters are highest, followed by Mature Singles
                    
                    3: €27.000<=33.500 (almost average), Empty nesters are highest, followed by Mature families

                    4: €33.500<=40.000 (average), 4: Empty Nesters are the highest, followed by Mature families
                    
                    5: €40.000<=67.000 (between 1 and 2 times average),Empty Nesters are highest followed by Mature families 
                        & Established families
                    
                    6: €67.000<=79.900 (2 times average), Empty Nesters are highest, followed by Established families
                    
                    
                    7: >79.900 (above 2 times average), Empty Nesters are highest, followed by Established families

                    
                    8: don't know / don't want to say, Empty Nesters are highest followed by Established families


                        



# 3. Effect of CICs on booking conversions:

```python
pd.read_sql('''select type_touch, purchase_any 
                from travel_data
                group by type_touch, purchase_any
                having type_touch between 7 and 15''',con=conn) #Between 7 & 15 the touch points are related to Travel Agencies
```

    Here we can see that in the first touch point, no purchases were made 
    
    But in the susequent touch point, a purchase has been made

# 4. Which are the consumer segments who engage in CICs and convert?

```python
#1 Considering Age, type of work & lifestages into consideration:

pd.read_sql('''select type_touch, count(type_touch) as touch_count, purchase_any , Age, BAS_werkzaamheid_resp, SPSS_Lifestage
                from travel_data join travel_data_demos on travel_data.userid = travel_data_demos.userid
                group by type_touch, BAS_werkzaamheid_resp, purchase_any, Age, SPSS_Lifestage
                having purchase_any =1
                order by touch_count desc limit 10''',con=conn) 
```

    Customer segments having highest touch points & conversion are of:
    
    Age 52, incapacitated & are Empty Nesters
    
    Touch point 1 is Accomodations Website
                7 is Touroperator / Travel agent Website Competitor
                10 is Touroperator / Travel agent Website Focus brand




```python
#2 Considering type of work & lifestages:

pd.read_sql('''select type_touch, count(type_touch) as touch_count, purchase_any , BAS_werkzaamheid_resp, SPSS_Lifestage
                from travel_data join travel_data_demos on travel_data.userid = travel_data_demos.userid
                group by type_touch, purchase_any, BAS_werkzaamheid_resp, SPSS_Lifestage
                having purchase_any =1
                order by touch_count desc limit 10''',con=conn)
```

    Customer segments having highest touch points & conversion are in:
    
    Early retirement & are Empty Nesters
    
    Touch point 1 is Accomodations Website
                7 is Touroperator / Travel agent Website Competitor


```python
#3 Considering only Age & Type of work:

pd.read_sql('''select type_touch, count(type_touch) as touch_count, purchase_any , Age, BAS_werkzaamheid_resp
                from travel_data join travel_data_demos on travel_data.userid = travel_data_demos.userid
                group by type_touch, BAS_werkzaamheid_resp, purchase_any, Age
                having purchase_any =1
                order by touch_count desc limit 10''',con=conn) 
```

```python
#4 Considering Age & Lifestage:

pd.read_sql('''select type_touch, count(type_touch) as touch_count, purchase_any , Age, SPSS_Lifestage
                from travel_data join travel_data_demos on travel_data.userid = travel_data_demos.userid
                group by type_touch, purchase_any, Age, SPSS_Lifestage
                having purchase_any =1
                order by touch_count desc limit 10''',con=conn)
```

# 5. Region,Municipality & Household size wise engagement in CICs 

```python
pd.read_sql('''select SPSS_Regio5, RESP_GEM_GROOTTE, BAS_huishoudgrootte,type_touch, count(type_touch) as touch_count
                from travel_data join travel_data_demos on travel_data.userid = travel_data_demos.userid
                group by type_touch, SPSS_Regio5, RESP_GEM_GROOTTE, BAS_huishoudgrootte
                
                order by touch_count desc limit 10''',con=conn)
```

    Region (SPSS_Regio5): 3.North, 1: Amsterdam, Rotterdam, Den Haag
    
    Size of manucipality(resp_gem_grootte): 4: 20.000-50.000  9: Amsterdam (>800.000)
    
    Size of Household(bas_huishoudgrootte): 1 to 3 people 

    Touch types: 1. Accomodation 7. Travel Agent Website Competitor 13. Flight tickets website
